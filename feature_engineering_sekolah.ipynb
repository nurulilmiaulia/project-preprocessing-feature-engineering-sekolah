```python
# STEP 4a: Lihat isi awal file Excel tanpa header
raw = pd.read_excel(data_path, header=None)
raw.head(15)
```

```python
# STEP 4b: Load ulang dataset dengan header yang benar (baris ke-2)
df = pd.read_excel(data_path, header=2)

# Cek hasil
df.head()
```

```python
# STEP 5: Rename kolom agar lebih sederhana & konsisten
df = df.rename(columns={
    "Siswa": "jumlah_siswa",
    "Mengulang": "mengulang",
    "Putus Sekolah": "putus_sekolah",
    "Kepala Sekolah & Guru": "kepala_guru",
    "Tenaga Kependidikan": "tendik",
    "Rombel": "rombel",
    "Ruang Kelas": "ruang_kelas",
    "Kecamatan": "kecamatan",
    "Sekolah": "nama_sekolah",
    "Status": "status_sekolah"
})

df.head()
```

```python
# STEP 6: Feature creation (pembuatan fitur baru)

# 1. Total guru
df["total_guru"] = df["kepala_guru"]

# 2. Rasio siswa per guru
df["rasio_siswa_guru"] = df["jumlah_siswa"] / df["total_guru"].replace(0, np.nan)

# 3. Rasio siswa per rombel
df["rasio_siswa_rombel"] = df["jumlah_siswa"] / df["rombel"].replace(0, np.nan)

# 4. Rasio siswa per ruang kelas
df["rasio_siswa_kelas"] = df["jumlah_siswa"] / df["ruang_kelas"].replace(0, np.nan)

# Tampilkan hasil
df[[
    "kecamatan",
    "jumlah_siswa",
    "total_guru",
    "rombel",
    "ruang_kelas",
    "rasio_siswa_guru",
    "rasio_siswa_rombel",
    "rasio_siswa_kelas"
]].head()
```

```python
# STEP 7: Membuat kategori kepadatan kelas

def kategori_kepadatan(x):
    if x < 20:
        return "Rendah"
    elif x < 30:
        return "Sedang"
    else:
        return "Tinggi"

df["kepadatan_kelas"] = df["rasio_siswa_kelas"].apply(kategori_kepadatan)

# Cek hasil
df[["kecamatan", "rasio_siswa_kelas", "kepadatan_kelas"]].head()
```

```python
# STEP 8: One-Hot Encoding untuk fitur kategorikal
categorical_cols = ["status_sekolah", "kepadatan_kelas", "kecamatan"]

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler

# STEP 9: Scaling fitur numerik
numeric_features = [
    "jumlah_siswa", "total_guru", "tendik",
    "rombel", "ruang_kelas",
    "rasio_siswa_guru", "rasio_siswa_rombel", "rasio_siswa_kelas"
]

scaler = StandardScaler()
df_encoded[numeric_features] = scaler.fit_transform(df_encoded[numeric_features])

df_encoded.head()
```

```python
# STEP 10: Simpan dataset akhir
output_path = "dataset_sekolah_feature_engineered.csv"
df_encoded.to_csv(output_path, index=False)

print("File tersimpan sebagai:", output_path)
```
